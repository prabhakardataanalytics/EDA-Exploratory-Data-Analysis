```python
Country, Invest1 = plt.subplots(figsize=(12, 7))
df2.groupby(['Select Investor1','Industry'])['Country'].value_counts().sort_values(ascending=False).head(5).plot(kind='barh')
plt.title('Top 5 Country Invested In Industry')
plt.xlabel('Investested  in Industry')
plt.ylabel('Country')
plt.show()
```

```python
max_valuations_by_industry = df2.groupby("Industry")["Valuation"].sum().sort_values(ascending=False).head(5)
max_valuations_by_industry
```

```python
max_valuations_by_industry = df2.groupby(['Industry'],as_index = False)['Valuation'].sum().sort_values(by = 'Valuation',ascending = False).head(5)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x = 'Industry', y = 'Valuation',data = max_valuations_by_industry)
```

**From the above graph we can see the high valuation is by Artificial Intelligence,E-commerce & direct-to-Consumer and other Industry**

```python
avg_valuations_by_industry = df2.groupby("Industry")["Valuation"].mean().sort_values(ascending=False).head(5)
avg_valuations_by_industry
```

```python
avg_valuation_by_industry_valuations_by_industry = df2.groupby(['Industry'],as_index = False)['Valuation'].mean().sort_values(by = 'Valuation',ascending = False).head(5)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x = 'Industry', y = 'Valuation',data = avg_valuation_by_industry_valuations_by_industry)
```
